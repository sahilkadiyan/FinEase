```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load your mixed data DataFrame
df = pd.read_csv('cleaned_data.csv')

# Select the columns with textual data for processing
textual_columns = ['Gender', 'Student','Married','Ethnicity','loan_intent','loan_grade','person_home_ownership','goals']  # Add the column names containing textual data

# Use CountVectorizer to transform text data into numerical features
vectorizer = CountVectorizer()
textual_features = vectorizer.fit_transform(df[textual_columns].astype(str))

# Standardize numerical features
numerical_columns = ['Gender', 'Student','Married','Ethnicity','loan_intent','loan_grade','person_home_ownership','goals']  # Add the numerical column names
numerical_features = df[numerical_columns]
scaler = StandardScaler()
numerical_features_standardized = scaler.fit_transform(numerical_features)

# Apply PCA for dimensionality reduction (optional)
n_components = 2  # Number of components after reduction
pca = PCA(n_components=n_components)
numerical_features_pca = pca.fit_transform(numerical_features_standardized)

# Combine numerical and transformed textual features
combined_features = pd.concat([pd.DataFrame(numerical_features_pca), pd.DataFrame(textual_features.toarray())], axis=1)

# Now you can use the combined_features for clustering

# Save the combined features DataFrame to a new CSV file
combined_features.to_csv('combined_features.csv', index=False)

print("Data transformed and saved as combined_features.csv.")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load your dataset, assuming it's stored in a DataFrame named 'data'
# Replace 'data.csv' with your actual dataset file
data = pd.read_csv('your_file_transformed.csv')

# Select the attributes for clustering
selected_attributes = ['Limit', 'Rating', 'Cards', 'Age', 'Education', 'Gender', 'Student', 'Married',
                       'person_home_ownership', 'person_emp_length', 'loan_intent', 'loan_grade', 'loan_amnt',
                       'loan_int_rate', 'loan_status', 'loan_percent_income', 'expenses', 'goals', 'Ethnicity',
                       'Balance', 'person_age', 'person_income']

# Extract the selected attributes
selected_data = data[selected_attributes]

# Convert categorical attributes to numerical using one-hot encoding
selected_data = pd.get_dummies(selected_data, columns=['Gender', 'Student', 'Married', 'person_home_ownership',
                                                       'loan_intent', 'loan_grade', 'Ethnicity'])

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data)

# Determine the optimal number of clusters using the elbow method
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow method results
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()

# Based on the elbow plot, choose the appropriate number of clusters
chosen_clusters = 3

# Perform K-Means clustering
kmeans = KMeans(n_clusters=chosen_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original DataFrame
data['Cluster'] = cluster_labels

# Explore the clusters and analyze patterns
cluster_summary = data.groupby('Cluster').mean()

# Display the summary of each cluster
print(cluster_summary)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load your dataset, replace 'data.csv' with your actual dataset file
data = pd.read_csv('newData2.csv')

# Define features for clustering
features = ['Limit', 'Rating', 'Cards', 'Age', 'Education', 'Gender', 'Student', 'Married', 'person_home_ownership',
            'person_emp_length', 'loan_intent', 'loan_grade', 'loan_amnt', 'loan_int_rate', 'loan_status',
            'loan_percent_income', 'expenses', 'goals', 'Ethnicity', 'Balance', 'person_age', 'person_income']

# Select the features for clustering
selected_data = data[features]

# Convert categorical features to numerical using one-hot encoding
selected_data_encoded = pd.get_dummies(selected_data, columns=['Gender', 'Student', 'Married', 'person_home_ownership',
                                                               'loan_intent', 'loan_grade', 'Ethnicity'])

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data_encoded)

# Perform K-Means clustering
chosen_clusters = 3  # Replace with the appropriate number of clusters
kmeans = KMeans(n_clusters=chosen_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original DataFrame
data['Cluster'] = cluster_labels

# Assign risk categories based on cluster characteristics
cluster_summary = data.groupby('Cluster').mean()

# Define the mapping of cluster to risk category
risk_mapping = {
    0: 'Low Risk',
    1: 'Medium Risk',
    2: 'High Risk'
}

# Add risk categories to the DataFrame
data['Risk Category'] = data['Cluster'].map(risk_mapping)

# Display the risk assessment results
risk_results = data[['Cluster', 'Risk Category']]

# Save the DataFrame with risk assessment to a new CSV file
data.to_csv('your_file_with_risk_categories.csv', index=False)

print("Risk assessment completed and data saved to 'your_file_with_risk_categories.csv'.")
```
